### Histogram plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].map(lambda x: iris.target_names[x])

# Set Seaborn style
sns.set(style='whitegrid')

# Plot distribution for each feature
for feature in iris.feature_names:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df, x=feature, hue='target_name', kde=True, palette='bright', bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

### Pair plot

```python
import pandas as pd
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].map(lambda x: iris.target_names[x])

# Create a pair plot
sns.set(style='whitegrid')
sns.pairplot(df, hue='target_name', palette='bright', diag_kind='kde')

plt.suptitle('Iris Dataset Pair Plot', y=1.02)
plt.show()
```

### Box plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].map(lambda x: iris.target_names[x])

# Set Seaborn style
sns.set(style='whitegrid')

# Create box plots for each feature
for feature in iris.feature_names:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='target_name', y=feature, hue='target_name', data=df, palette='pastel', dodge=False, legend=False)
    plt.title(f'Box Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

### Violin plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].map(lambda x: iris.target_names[x])

# Set Seaborn style
sns.set(style='whitegrid')

# Violin plot with hue set
for feature in iris.feature_names:
    plt.figure(figsize=(8, 5))
    sns.violinplot(
        x='target_name', 
        y=feature, 
        hue='target_name',       # explicitly set hue
        data=df, 
        palette='muted', 
        dodge=False,             # prevents duplication of violins
        legend=False             # avoid redundant legend
    )
    plt.title(f'Violin Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

### Correlation plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, vmin=-1, vmax=1)
plt.title('Correlation Matrix of Iris Features')
plt.tight_layout()
plt.show()
```
